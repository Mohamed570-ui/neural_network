```python
import matplotlib.pyplot as plt

# Exemple de donnes simules (remplace par tes vraies donnes)
epochs = range(1, 101)  # 100 poques

# Simuler la perte dcroissante pour l'entranement et la validation
loss_train = [1.0 / (epoch**0.5) + 0.05 * (0.5 - epoch % 2) for epoch in epochs]
loss_val = [1.1 / (epoch**0.5) + 0.05 * (epoch % 3 - 1) for epoch in epochs]

# Trac des courbes
plt.figure(figsize=(8,5))
plt.plot(epochs, loss_train, label='Perte Entranement', color='blue', linewidth=2)
plt.plot(epochs, loss_val, label='Perte Validation', color='orange', linewidth=2)

plt.title("Courbe de la perte pendant l'entranement et la validation", fontsize=14)
plt.xlabel("poques", fontsize=12)
plt.ylabel("Perte", fontsize=12)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Simuler les donnes de prcision ( remplacer par tes vraies donnes)
epochs = range(1, 101)

# Simulations ralistes de prcision croissante
accuracy_train = [0.5 + 0.005 * epoch - 0.01 * (epoch % 5 == 0) for epoch in epochs]
accuracy_val = [0.48 + 0.0045 * epoch - 0.015 * (epoch % 7 == 0) for epoch in epochs]

# Trac des courbes
plt.figure(figsize=(8,5))
plt.plot(epochs, accuracy_train, label="Prcision Entranement", color="green", linewidth=2)
plt.plot(epochs, accuracy_val, label="Prcision Validation", color="red", linewidth=2)

plt.title("Courbe de prcision pendant l'entranement et la validation", fontsize=14)
plt.xlabel("poques", fontsize=12)
plt.ylabel("Prcision", fontsize=12)
plt.ylim(0, 1)  # Limite entre 0 et 1 pour une mtrique de prcision
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
import numpy as np

# ---  remplacer par tes vraies tiquettes ---
# y_true : vraies classes
# y_pred : classes prdites par ton modle
y_true = [0, 1, 2, 2, 1, 0, 1, 2, 0, 1] * 10  # Exemple
y_pred = [0, 2, 2, 2, 1, 0, 1, 0, 0, 1] * 10  # Exemple

# Calcul de la matrice de confusion
cm = confusion_matrix(y_true, y_pred)
classes = [f"Classe {i}" for i in range(cm.shape[0])]  #  adapter si tu as des noms spcifiques

# Affichage avec seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=classes, yticklabels=classes)

plt.title("Matrice de confusion (validation/test)", fontsize=14)
plt.xlabel("Classe prdite", fontsize=12)
plt.ylabel("Classe relle", fontsize=12)
plt.tight_layout()
plt.show()
```
